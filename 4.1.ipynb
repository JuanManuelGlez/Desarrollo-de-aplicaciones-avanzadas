```python
from fitter import Fitter

# Datos hipotticos de ventas diarias
ventas_diarias = [50, 60, 58, 52, 49, 45, 55, 61, 57, 62, 53, 64, 56, 50, 48, 52, 55, 60, 49, 53, 
                  59, 61, 57, 62, 58, 60, 53, 55, 51, 54]

# Ajustar los datos a diferentes distribuciones
f = Fitter(ventas_diarias, distributions=['norm', 'expon', 'gamma', 'lognorm', 'beta'])
f.fit()

# Mostrar el resumen de las mejores distribuciones
f.summary()

```
